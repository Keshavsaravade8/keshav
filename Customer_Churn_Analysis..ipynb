

# Data Visualization

```python

total_churn_person = sns.countplot(x="Churn", data=data)
total_churn_person.bar_label(total_churn_person.containers[0])
plt.title("Total Number of Churned vs Non-Churned Customers")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```

```python
##same graph more things add
# Set a pleasant visual theme
sns.set_theme(style="whitegrid")

# Set figure size
plt.figure(figsize=(8, 6))

# Custom colors for bars
custom_palette = ["#66c2a5", "#fc8d62"]  # Greenish and orangish

# Create the count plot
total_churn_person = sns.countplot(x="Churn", data=data, palette=custom_palette)

# Add value labels on top of bars
total_churn_person.bar_label(total_churn_person.containers[0], padding=5, fontsize=12, weight='bold')

# Add chart title and axis labels with formatting
plt.title("📊 Customer Churn Overview", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("Churn Status", fontsize=12, fontweight='bold')
plt.ylabel("Number of Customers", fontsize=12, fontweight='bold')

# Customize tick parameters
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Add grid for Y axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

 Insight:The bar chart reveals a significant churn rate among customers, highlighting the need for deeper investigation
        into churn drivers.


```python
# Group by 'Churn' and count occurrences
pie_chart = data['Churn'].value_counts()

# Labels and values
labels = pie_chart.index
sizes = pie_chart.values

# Plotting the pie chart
plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=140)
plt.title("📊 Customer Churn Distribution", fontsize=16, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures a perfect circle
plt.show()
```

Insight:The pie chart shows that approximately a quarter of customers have churned, emphasizing the importance of customer 
    retention strategies.


```python
total_genderwise = sns.countplot(x="gender", data=data)
total_genderwise.bar_label(total_genderwise.containers[0])
plt.title("Total Number of Male vs Female")
plt.xlabel("gender")
plt.ylabel("Number of Customers")

plt.show()
```

```python
# Set a clean, appealing theme
sns.set_theme(style="whitegrid")

# Set figure size
plt.figure(figsize=(8, 6))

# Define custom colors (e.g., blue for male, pink for female)
custom_palette = ["#66c2a5", "#fc8d62"]  # Or try ['#4c72b0', '#dd8452']

# Create count plot
total_genderwise = sns.countplot(
    x="gender",
    data=data,
    palette=custom_palette,hue="Churn"
)

# Add value labels on top of bars
total_genderwise.bar_label(
    total_genderwise.containers[0],
    padding=5,
    fontsize=12,
    weight='bold'
)

# Add chart title and axis labels with formatting
plt.title("👥 Gender-wise Customer Distribution", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("Gender", fontsize=12, fontweight='bold')
plt.ylabel("Number of Customers", fontsize=12, fontweight='bold')

# Customize tick parameters
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Add Y-axis grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Final layout and display
plt.tight_layout()
plt.show()
```

Insight:Both male and female customers show similar churn patterns, indicating that gender has minimal impact on churn behavior.


```python
# Apply Seaborn theme
sns.set_theme(style="whitegrid")

# Count values for 'gender'
gender_counts = data['gender'].value_counts()
labels = gender_counts.index
sizes = gender_counts.values

# Define custom colors (same as previous bar chart)
colors = ["#66c2a5", "#fc8d62"]

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    textprops={'fontsize': 12, 'weight': 'bold'}
)

# Optional: Add donut-style center
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

# Title
plt.title("👥 Gender-wise Customer Distribution (in %)", fontsize=16, fontweight='bold', pad=20)
plt.axis('equal')  # Keeps pie chart circular
plt.tight_layout()

# Show plot
plt.show()
```

```python
# List of service-related features
features = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Apply Seaborn theme
sns.set_theme(style="whitegrid")

# Create subplots (3 rows x 3 cols)
fig, axes = plt.subplots(3, 3, figsize=(20, 14))
axes = axes.flatten()

# Define a color palette for Churn
palette = {"Yes": "#fc8d62", "No": "#66c2a5"}

# Loop through features and plot each
for i, feature in enumerate(features):
    ax = axes[i]
    plot = sns.countplot(
        x=feature,
        data=data,
        hue="Churn",
        palette=palette,
        ax=ax
    )

    # Add value labels for each hue (Yes/No)
    for container in plot.containers:
        plot.bar_label(container, padding=3, fontsize=9, weight='bold')

    # Set title and remove axis labels for cleaner look
    ax.set_title(f"{feature} vs Churn", fontsize=13, fontweight='bold')
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.tick_params(axis='x', rotation=15)

# Add a legend just once outside the plots
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, title='Churn', loc='upper center', ncol=2, fontsize=11, title_fontsize=12)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.suptitle("📊 Service Feature vs Churn Analysis", fontsize=18, fontweight='bold', y=1.02)

# Show the full plot
plt.show()
```

Churn by Services Insight:Customers not subscribed to security, backup, or tech support services exhibit significantly
    higher churn rates, highlighting service value in retention.

# Summary

The data shows a high churn rate, with around a quarter of customers leaving. Churn is similar across genders, and the customer base is almost evenly split between males and females. Most customers are younger, as there are no senior citizens. Churn is higher among newer customers (0-10 months) and lower among long-term customers (60+ months). Month-to-month contracts have the highest churn, while one- or two-year contracts have better retention.
Customers without services like Online Security, Tech Support, and Device Protection are more likely to churn. Additionally, customers using electronic checks have the highest churn, while those using automatic payments (bank/credit card) are more likely to stay.
This suggests that improving service features and offering longer contracts could help reduce churn.
